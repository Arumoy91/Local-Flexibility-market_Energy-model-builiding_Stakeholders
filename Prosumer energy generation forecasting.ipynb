```python
from sklearn.preprocessing import LabelEncoder

def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python
df = df.apply(lambda x: object_to_int(x))
df.head()
```

```python
sns.distplot(df.Total_Heating_Energy)
```

```python
y = df.Photovoltaic_Power
X = df.drop(["Photovoltaic_Power"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=30,random_state=10)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
```

```python
ln.fit(X_train,y_train)
```

```python
training_data_prediction = ln.predict(X_train)
training_data_prediction
```

```python
y_pred=ln.predict(X_test)
y_pred
```

```python
ln.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
print("R-Squared:",r2_score(y_test,y_pred))
```

```python
print( ln.intercept_ )
```

```python
#get slope
print( ln.coef_ )
```

```python
import numpy as np

residuals = y_test - y_pred

# Calculate the degrees of freedom
n = len(y_test)
p = 1  # Assuming you have one predictor variable in your linear regression model
df = n - p - 1  # n is the number of samples, p is the number of features

# Calculate the sum of squared residuals
sum_squared_residuals = np.sum(residuals ** 2)

# Calculate the standard error
standard_error = np.sqrt(sum_squared_residuals / df)

# Print the result
print("Standard Error of Residuals:", standard_error)
```

```python
from sklearn.metrics import mean_absolute_error


mae = mean_absolute_error(y_test, y_pred)

# Print the result
print("Mean Absolute Error (MAE):", mae)
```
